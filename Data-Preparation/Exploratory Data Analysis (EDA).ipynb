```python
import seaborn as sns
import matplotlib.pyplot as plt

attributes = ['Calories', 'FatContent', 'SaturatedFatContent', 'CholesterolContent',
              'SodiumContent', 'CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent']

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))

for i, attribute in enumerate(attributes):
    row = i // 3
    col = i % 3
    sns.histplot(data[attribute], bins=range(0, int(data[attribute].max()) + 500, 500), kde=False, ax=axes[row, col])
    axes[row, col].set_title(f'Histogram of {attribute}')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Assuming 'data' is your DataFrame
attributes = ['Calories', 'FatContent', 'SaturatedFatContent', 'CholesterolContent',
              'SodiumContent', 'CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent']

height_values = []  # To store the height values

for attribute in attributes:
    heights, bins = np.histogram(data[attribute], bins=range(0, int(data[attribute].max()) + 500, 500), density=False)
    height_values.append(heights)  # Append the heights for each attribute

# 'height_values' now contains the heights for each column's histogram
for i, attribute in enumerate(attributes):
    print(f'\nAttribute: {attribute}')
    print('Bin Value\tCount')
    for bin_value, height_value in zip(bins, height_values[i]):
        print(f'{bin_value}\t\t{height_value}')
```

```python
column_names = ['Calories', 'FatContent', 'SaturatedFatContent', 'CholesterolContent', 'SodiumContent', 'CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent']

for column in column_names:
    counts = data[column].value_counts()
    print(f"Counts for {column}:\n{counts}\n")
```

```python
calories_counts = data['Calories'].value_counts()
print("Calories counts:\n", calories_counts)

fat_content_counts = data['FatContent'].value_counts()
print("FatContent counts:\n", fat_content_counts)
```
